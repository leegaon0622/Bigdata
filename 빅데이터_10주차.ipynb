2. class 별로 승객 수를 나타내는 countplot을 구하시오. (생존 여부('alive)'를 세부 분류로 함.)

```python
import seaborn as sns
titanic=sns.load_dataset('titanic')
sns.countplot(data=titanic,x='class',hue='alive')
```

3. class 별 count를 이용하여 pie chart를 작성 하시오.(image 참조)
a-3.png

```python
import matplotlib.pyplot as plt
import seaborn as sns
titanic=sns.load_dataset('titanic')
result=titanic.groupby('class')['class'].count()
labels = ['first', 'second', 'third']
plt.pie(result,labels=labels, autopct='%.1f%%')
plt.show()

```

6.결혼 유무에 따른 파이 차트를 구하시오

```python
import seaborn as sns
result=df.groupby('결혼 유무')['결혼 유무'].count()
print(result)
```

```python
import matplotlib.pyplot as plt
labels = ['결혼', '무응답', '이혼혼']
plt.pie(result,labels=labels, autopct='%.1f%%')
plt.show()
```
