의사 결정 트리 분류

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

df = DecisionTreeClassifier(max_depth=3, random_state=42)
df.fit(X, y)

cscore=cross_val_score(df,X,y,cv=5)  # 교차 검증 k=5
print('accuracy',cscore.mean())
X.shape
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(df)
plt.show()
```

의사 결정 트리 회귀

```python
# 2차식으로 만든 데이터셋 + 잡음
import numpy as np
np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10
```

```python
plt.scatter(X,y)
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import cross_val_score
y=y.squeeze() ##y가 1차원일때는 삭제
tree_reg = DecisionTreeRegressor(max_depth=3, random_state=42)
tree_reg.fit(X, y)
cscore=cross_val_score(tree_reg,X,y,cv=5,scoring='neg_mean_squared_error')
print('RMSE',np.sqrt(-cscore.mean()))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
plt.figure(figsize=(10,7))
plot_tree(tree_reg  )
plt.show()
```

랜덤 포레스트 분류

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target

df = RandomForestClassifier(n_estimators=30)
df.fit(X, y)

cscore=cross_val_score(df,X,y,cv=5)
print('accuracy',cscore.mean())
y.shape
#df.predict([[0.1,0.2]]) 새로운 입력에 대한 예측
```

```python
# 2차식으로 만든 데이터셋 + 잡음
import numpy as np
np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
y = y + np.random.randn(m, 1) / 10
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
y=y.squeeze() #y가 1차원일때는 삭제
tree_reg = RandomForestRegressor(n_estimators=50)
tree_reg.fit(X, y)
cscore=cross_val_score(tree_reg,X,y,cv=5,scoring='neg_mean_squared_error')
print('RMSE',np.sqrt(-cscore.mean()))
X.shape
print(y.shape)
```
