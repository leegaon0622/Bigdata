과제 2. tip 데이터를 이용하여 그림과 같이 subplot을 나타내시오.  
1) bar chart (x: total bill, y: tips)  
2) scatter chart (x: total bill, y: tips) ; 하위 분류 :day  
3) count chart (요일별)

sub.png

```python
%reset -f
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('white')
tips=sns.load_dataset("tips")
df=pd.DataFrame(tips)
print(df.head())
print("\n")

fig=plt.figure(figsize=(20,10))
ax1=fig.add_subplot(2,3,1)
ax2=fig.add_subplot(2,3,2)
ax3=fig.add_subplot(2,3,3)

#1)
sns.barplot(data=tips,x='day',y='total_bill', ax=ax1)

#2)
sns.scatterplot(data=tips,x='total_bill',y='tip', hue='day',ax=ax2)

#3)
sns.countplot(data=tips,x='day',ax=ax3)
```

```python
## 3-1
import pandas as pd

df=pd.read_csv('/content/contents2.csv')

import plotly.express as px
import matplotlib.pyplot as plt
print(df.head())

fig=px.bar(df, x='month', y='음악',color='month', color_continuous_scale=px.colors.diverging.Spectral)
fig.update_layout(width=600)


```

```python
## 3-2
import pandas as pd

df=pd.read_csv('/content/contents2.csv')

import plotly.express as px
import matplotlib.pyplot as plt
print(df.head())

fig=px.bar(df, x='month', y='음악',color='month', color_continuous_scale=px.colors.sequential.Purples)
fig.update_layout(width=600)

fig.write_html('/content/3-2.html')
```

과제4. 아래 데이터를 사용하여 레이더 차트(line poral)
를 만드는 code를 작성 하시오.   

```python
import plotly.express as px

df=pd.read_csv('/content/score.csv')
vals=df.iloc[3,1:].values
print(vals)
col=['국어','영어','수학','과학','음악','체육']

fig = px.line_polar(df, r=vals,theta=col,line_close=True)
fig.update_traces(fill='toself')
```
