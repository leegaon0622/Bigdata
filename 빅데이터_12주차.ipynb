과제1) 인천광역시 미추홀구_재래시장_20220822.csv 화일에 포함되어 있는   
재래시장들의 위치를 marker에 표시 하시오.   
 지도의 중심 위치는 7호관뒤 수준 원점으로 설정 하시오   
 marker를 클릭 했을때 시장 이름이 나타나도록 설정     

```python
# -*- coding: utf-8 -*-

# 라이브러리 불러오기
import pandas as pd
import folium

#  데이터프레임 변환
df = pd.read_csv('/content/인천광역시 미추홀구_재래시장_20220822.csv',  encoding = 'CP949')  # 수준원점 기준

print(df)

# 지도 만들기
# 7호관뒤 수준 원점 중심 37.449405, 126.657369
smap = folium.Map(location=[37.449405, 126.657369], tiles='Stamen Terrain', zoom_start=14)


# 재래시장 위치정보를 Marker로 표시
for name, lat, lng in zip(df.시장명, df.위도, df.경도):
  folium.Marker([lat, lng], popup=name).add_to(smap)
smap

```

과제2) 서울특별시 성북구_약국list.csv 화일에 포함되어 있는   
약국들의 위치를 circlemarker로 표시 하시오.  
 지도의 중심 위치는 성북 구청으로 설정 하시오   
 circlemarker를 클릭 했을때 약국 이름이 나타나도록 설정

```python
# -*- coding: utf-8 -*-

# 라이브러리 불러오기
import pandas as pd
import folium

#  리스트를 데이터프레임 변환
df = pd.read_csv('/content/서울특별시 성북구_약국list.csv',  encoding = 'CP949')

# 지도 만들기
# 성북 구청 37.589375 , 127.016741
smap = folium.Map(location=[37.589375 , 127.016741], tiles='Stamen Terrain', zoom_start=14)

for name, lat, lng in zip(df.약국명, df.위도, df.경도):
  folium.CircleMarker([lat, lng],radius=10,color='brown', fill=True,fill_color='coral',
      fill_opacity=0.7,popup=name).add_to(smap)

smap
```

과제3. 서울시의 구별 인구수 ( 총생활인구수    )를 단계 구분도(choropleth)를 이용하여 표시 하시오.       
구분이 잘 되도록 threshold_scale을 적절히 설정.  

chprpleth 함수의 parameter 중 key_on은
'feature.properties.name' 으로 설정.


```python
import pandas as pd

import folium

df=pd.read_csv('/content/서울구별인구.csv')
#df.columns = df.columns.map(str)


print(df.head())
import requests
import json
import re

# 서울 행정구역 json raw파일(githubcontent)https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json
s_geo='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'

s_map = folium.Map(location=[37.559984,126.9753071],  # 숭례문 위도, 경도
                   tiles='Stamen Terrain', zoom_start=11)


s_map

# Choropleth 클래스로 단계구분도 표시하기
folium.Choropleth(geo_data=s_geo, data = df,columns = ['시군구명','인구수'],
fill_color='YlOrRd', fill_opacity=0.7, line_opacity=0.3,
threshold_scale=[200, 300, 400, 500, 600, 700, 800],
key_on='feature.properties.name').add_to(s_map)

s_map
```
