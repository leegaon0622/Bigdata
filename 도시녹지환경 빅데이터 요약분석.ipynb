```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

filename = '전국도시공원정보표준데이터.csv'
df_park = pd.read_csv(filename, encoding='cp949')
```

```python
#앞부분 미리보기
df_park.head()
```

```python
#데이터프레임의 구조
df_park.info() 
```

```python
#데이터프레임의 데이터 개수 확인
df_park.count()
```

```python
# 공원면적의 평균값
print('공원면적의 평균')
print(df_park['공원면적'].mean()) 
print('\n')

# 공원면적의 최대값
print('공원면적의 최대값')
print(df_park['공원면적'].max()) 
print('\n')

# 공원면적의 최소값 
print('공원면적의 최소값')
print(df_park['공원면적'].min()) 

```

```python
#공원면적 Null값 찾기
missing_df = df_park.isnull()
for col in missing_df.columns:
  missing_count = missing_df[col].value_counts() # 각 열의 NaN 개수 파악
  try:
     print(col, ':', missing_count[True]) # NaN 값이 있으면 개수 출력
  except:
     print(col,':',0) # NaN 값이 없으면 0개 출력
```

```python
#도로명 주소 null값의 수
print(df_park['소재지도로명주소'].isnull().sum())

#지번 주소 null값의 수
print(df_park['소재지지번주소'].isnull().sum())

#'도로명 주소'와 ‘지번 주소’가 모두 결측치가 아닌 경우
df_park[df_park['소재지도로명주소'].notnull() & df_park['소재지지번주소'].notnull()].shape
```

```python
#공원구분 중복데이터 처리
df_park2 = df_park.drop_duplicates(subset=['공원구분']) 
print(df_park2['공원구분'])
```

```python
#각 지역구 별 공원 수
df_park["si"] = df_park['소재지도로명주소'].str.split(expand=True)[0]
df_park["si"].value_counts()
```

```python
#지역 통일
df_park["si"] = df_park["si"].replace("영월군", "강원도")
df_park["si"] = df_park["si"].replace("대전", "대전광역시")
df_park["si"] = df_park["si"].replace("서울", "서울특별시")
df_park["si"].value_counts()
```

```python
#위도 경도를 이용한 scatterplot
import pandas as pd
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt

sns.scatterplot(data = df_park,x='경도', y='위도', color='Green')
```

```python
#범위를 벗어난 위도 경도 재조정 후 scatterplot
df = df_park[(df_park["경도"]<132)&(df_park["위도"]>32)].copy()
sns.scatterplot(data = df,x='경도', y='위도',color='Green')
```

```python
#공원구분 분포도 scatterplot
plt.figure(figsize=(8,9))
sns.scatterplot(data = df,x='경도', y='위도', hue="공원구분", sizes=(10,100))
```

```python
#시별 분포도 scatterplot
plt.figure(figsize=(8,9))
sns.scatterplot(data = df,x='경도', y='위도', hue="si", sizes=(10,100))
```

```python
#년도별 공원 수 histplot으로 보기
#지정고시일 데이터 타입 바꾸기
df_park["지정고시일"] = pd.to_datetime(df_park["지정고시일"])

#지정고시일 년으로 나누기
df_park["year"] = df_park["지정고시일"].dt.year

sns.histplot(df_park["year"], color='Green')
```
